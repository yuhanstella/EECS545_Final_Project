## Change 'categorical' key and mode to One-Hot Encoding

### Change key from data.csv

```python
df.dropna(inplace=True)
df['key'] = df['key'].replace({'A': 0, 'A#': 1, 'B': 2, 'C': 3, 'C#': 4, 'D': 5,
                               'D#': 6, 'E': 7, 'F': 8, 'F#': 9, 'G': 10, 'G#': 11})
print(df['music_genre'].unique())
df['music_genre'] = df['music_genre'].replace(
    {'Electronic': 0, 'Anime': 1, 'Jazz': 2, 'Alternative': 3, 'Country': 4, 'Rap': 5, 'Blues': 6, 'Rock': 7, 'Classical': 8, 'Hip-Hop': 9})
df_key = df[['key']]
df_genre = df[['music_genre']]
df_key_onehot = OneHotEncoder(dtype=np.int8).fit_transform(df_key).toarray()
df_genre_onehot = OneHotEncoder(
    dtype=np.int8).fit_transform(df_genre).toarray()
```

```python
df.drop(['key'], axis=1, inplace=True)
df.drop(['instance_id'], axis=1, inplace=True)
df.drop(['artist_name'], axis=1, inplace=True)
df.drop(['track_name'], axis=1, inplace=True)
df.drop(['obtained_date'], axis=1, inplace=True)
key_indices = [f'key_{i}' for i in range(12)]
df.drop(['music_genre'], axis=1, inplace=True)
genre_indices = [f'genre_{i}' for i in range(10)]
print(key_indices)
```

```python
print(df_key_onehot.shape)
df = df.join(pd.DataFrame(df_key_onehot, columns=key_indices))
df = df.join(pd.DataFrame(df_genre_onehot, columns=genre_indices))
```

```python
df.head()
```

```python
df_mode = df[['mode']]
df_mode_onehot = OneHotEncoder(dtype=np.int8).fit_transform(df_mode).toarray()
```

```python
df.drop(['mode'], axis=1, inplace=True)
df = df.join(pd.DataFrame(df_mode_onehot, columns=['Minor', 'Major']))
```
