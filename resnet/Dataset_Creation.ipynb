## Read various CSV files

```python
df = pd.read_csv(
    '/content/drive/MyDrive/Colab Notebooks/project/music_genre.csv')
df.replace(np.inf, 0)
df.replace(np.nan, 0)
df.replace(-np.inf, 0)
df.dropna(subset='popularity')
df.round({'popularity': 0 })
```

## Change 'categorical' key and mode to One-Hot Encoding

### Change key from data.csv

```python
df.dropna(inplace=True)
```

```python
df.drop(['key'], axis=1, inplace=True)
df.drop(['instance_id'], axis=1, inplace=True)
df.drop(['artist_name'], axis=1, inplace=True)
df.drop(['acousticness'], axis=1, inplace=True)

# df.drop(['track_name'], axis=1, inplace=True)
df.drop(['obtained_date'], axis=1, inplace=True)
# key_indices = [f'key_{i}' for i in range(12)]
df.drop(['music_genre'], axis=1, inplace=True)
df.drop(['danceability'], axis=1, inplace=True)
df.drop(['duration_ms'], axis=1, inplace=True)
df.drop(['energy'], axis=1, inplace=True)
df.drop(['instrumentalness'], axis=1, inplace=True)
df.drop(['liveness'], axis=1, inplace=True)
df.drop(['loudness'], axis=1, inplace=True)
df.drop(['mode'], axis=1, inplace=True)
df.drop(['valence'], axis=1, inplace=True)
df.drop(['speechiness'], axis=1, inplace=True)
df.drop(['tempo'], axis=1, inplace=True)



df['popularity'][df['popularity'] <= 34] = 0
df['popularity'][(df['popularity'] > 34) & (df['popularity'] <= 45)] = 1
df['popularity'][(df['popularity'] > 45) & (df['popularity'] <= 56)] = 2
df['popularity'][df['popularity'] > 56] = 3





# genre_indices = [f'genre_{i}' for i in range(10)]
# print(key_indices)
```
